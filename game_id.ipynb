```python
def write_to_csv(data, game_id, y, m, d):
    with open("/Users/heng/Data_Base/project/Game_data/" + str(y) + "-" + str(m) + "-" + str(d) + ":" + str(game_id - 1), 'a') as outcsv:   
        print(str(y) + "-" + str(m) + "-" + str(d) + ":" + str(game_id-1))
        writer = csv.writer(outcsv, delimiter=',', quotechar='|', quoting=csv.QUOTE_MINIMAL, lineterminator='\n')
        writer.writerow(['Game_ID', 'Away_No', 'Away_Point', 'Host_No', 'Host_Point'])
        writer.writerow([data[0], data[1], data[2], data[3], data[4]])
```
