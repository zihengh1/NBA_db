```python
team_id = {"BOS": "0", "BRK": "1", "NYK": "2", "PHI": "3", "TOR": "4", \
           "CHI": "5", "CLE": "6", "DET": "7", "IND": "8", "MIL": "9", \
           "ATL": "10", "CHO": "11", "MIA": "12", "ORL": "13", "WAS": "14", \
           "DEN": "15", "MIN": "16", "POR": "17", "OKC": "18", "UTA": "19", \
           "GSW": "20", "LAC": "21", "LAL": "22", "PHO": "23", "SAC": "24", \
           "DAL": "25", "HOU": "26", "MEM": "27", "NOP": "28", "SAS": "29"}
```

```python
def deal_with_file_name(filename):
    filename = filename.split("/")
    short_name = filename[-1][0:9]
    short_name = short_name.replace(".", "")
    if short_name[-1].isdigit() == False:
        short_name = short_name[:-1]
    # print(short_name)
    player_name = ""
    p_id = -1
    for R_name, S_name in table_dict.items():
        if S_name == short_name:
            player_name = R_name
            for Player_ID, p_n in player_dict.items():
                if player_name == p_n:
                    p_id = Player_ID
    # print(p_id)
    # print(player_name)
    if p_id != -1:
        # print("got it")
        return player_name, p_id
    else:
        return player_name, p_id
```

```python
path = "/Users/heng/Desktop/regular_season/"
for filename in glob.glob(os.path.join(path, '*.csv')):
    data = pd.read_csv(filename)
    (player_name, ID) = deal_with_file_name(filename)
    if ID != -1:
        print(player_name)
        data = data.loc[:, ['G', 'Date', 'Tm', 'Opp', 'MP', 'FG%', '3P%', 'TRB', 'AST', 'STL', 'BLK', 'PTS', '+/-']]
        data = data[pd.notnull(data['Date'])]
        data['G'] = data['G'].fillna('no play')
        data['Tm'].replace(team_id, inplace=True)
        data['Opp'].replace(team_id, inplace=True)
        data.insert(0, "Player_ID", str(ID))
        for index, row in data.iterrows():
            d = row["Date"]
            time = d.split("-")
            string_time = str(int(time[0])) + "-" + str(int(time[1])) + "-" + str(int(time[2]))
            Tm = int(row["Tm"])
            Opp = int(row["Opp"])
            for key, value in game_dict.items():
                if string_time == value[2]:
                    if (Tm == value[1] and Opp == value[0]):
                        data.loc[index,"Game_ID"]= str(key)
                    elif (Tm == value[0] and Opp == value[1]):
                        data.loc[index,"Game_ID"]= str(key)
        data = data.drop(columns = "Tm")
        data = data.drop(columns = "Opp")
        data = data.drop(columns = "Date")
        data.to_csv('/Users/heng/Data_Base/project/Performance_data/' + player_name, index = False)
```
